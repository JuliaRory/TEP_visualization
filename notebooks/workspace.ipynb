```python
import bisect
from collections import deque
import copy

class RollingMedian:
    def __init__(self, data, n_max):
        self.n = n_max
        self.window = deque(data[:n_max])
        self.sorted_window = copy.deepcopy(data[:n_max])
        self.sorted_window.sort()

    def add(self, value):
        # добавляем новое значение
        bisect.insort(self.sorted_window, value)
        self.window.append(value)

        # если превысили размер окна — удаляем самое старое
        if len(self.window) > self.n:
            old = self.window.popleft()
            self.sorted_window.remove(old)

    def median(self):
        m = len(self.sorted_window)
        if m == 0:
            return None
        if m % 2 == 1:
            return self.sorted_window[m // 2]
        else:
            return (self.sorted_window[m // 2 - 1] + self.sorted_window[m // 2]) / 2
```

```python
def apply_car_matrix(data):
    """
    Реализация CAR как пространственного фильтра (матрица умножения).
    data : np.ndarray, shape (n_channels, n_samples)
    """

    n_channels = data.shape[0]
    I = np.eye(n_channels)
    ones = np.ones((n_channels, n_channels)) / n_channels
    W = I - ones  # матрица фильтра
    
    return W @ data  
```

```python
from collections import deque

class RollingMean:
    def __init__(self, data, n_max, save_all=False):
        self.n = n_max
        self.trim_last = not save_all
        self.window = deque(data[:n_max], maxlen=n_max if self.trim_last else None)
        self.sum = sum(self.window)

    def add(self, value):
        if len(self.window) == self.n and self.trim_last:
            old = self.window.popleft()
            self.sum -= old
        self.window.append(value)
        self.sum += value

    def calculate(self):
        if len(self.window) == 0:
            return None
        return round(self.sum / len(self.window), 2)
```

```python
rm = RollingMean([2, 3], n_max=3)

print(rm.calculate())  # 2.0

rm.add(4)
print(rm.calculate())  # (2+3+4)/3 = 3.0

rm.add(5)
print(rm.calculate())  # (3+4+5)/3 = 4.0
```

```python
from scipy.ndimage import uniform_filter1d

def remove_and_interpolate(data, start, end, kernel_size=5):
    # data : np.ndarray Массив формы (n_channels, n_samples).
    
    data = data.copy()
    n_ch, n_samples = data.shape
    
    # индексы концов
    x = np.array([start-1, end+1])
    if x[0] < 0: x[0] = 0
    if x[1] >= n_samples: x[1] = n_samples-1
    
    for ch in range(n_ch):
        # значения на концах
        y = np.array([data[ch, x[0]], data[ch, x[1]]])
        
        # линейная интерполяция
        interp = np.interp(np.arange(start, end+1), x, y)
        data[ch, start:end+1] = interp
        
        # сглаживание окрестностей границ
        left = max(0, start - kernel_size)
        right = min(n_samples, end + kernel_size + 1)
        data[ch, left:right] = uniform_filter1d(data[ch, left:right], size=kernel_size)
    
    return data
```

```python
def resample(data, old_rate, new_rate):

    # data : np.ndarray  Исходный массив формы (n_channels, n_samples).

    n_channels, n_samples = data.shape
    duration = n_samples / old_rate  # общая длина в секундах
    new_samples = int(round(duration * new_rate))

    # временные оси
    t_old = np.linspace(0, duration, n_samples, endpoint=False)
    t_new = np.linspace(0, duration, new_samples, endpoint=False)

    # интерполяция по каждому каналу
    resampled = np.empty((n_channels, new_samples), dtype=float)
    for ch in range(n_channels):
        resampled[ch] = np.interp(t_new, t_old, data[ch])

    return resampled
```
